```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
#matplotlib inline
import pandas as pd
```

```python
import sklearn.linear_model as lm
from sklearn.metrics import r2_score
```

```python
np.random.seed(42)
n_samples = 100

X = np.linspace(0, 10, 100)
rng = np.random.randn(n_samples) * 100
y = X ** 3 + rng + 100

plt.figure(figsize=(10,8));
plt.scatter(X, y);
```

```python
or_lr = lm.LinearRegression()
or_lr.fit(X.reshape(-1,1),y)
y_pred = or_lr.predict(X.reshape(-1,1))

plt.figure(figsize=(15,12))
plt.plot(X,y_pred)
plt.scatter(X,y)
print("R^2 Score is:",r2_score(y,y_pred))
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures()
X_poly = poly_reg.fit_transform(X.reshape(-1, 1))

or_lr_2 = or_lr.fit(X_poly,y)
y_pred = or_lr.predict(X_poly)

plt.figure(figsize=(15,12))
plt.plot(X,y_pred)
plt.scatter(X,y)
print("R^2 Score is:",r2_score(y,y_pred))
plt.show()
```

```python
from sklearn.datasets import load_boston
Boston_data = load_boston()
df = pd.DataFrame(Boston_data.data,columns=Boston_data.feature_names)
df.head()
```

```python
X = df['DIS'].values
y = df['NOX'].values
```

```python
or_lr = lm.LinearRegression()
or_lr.fit(X.reshape(-1,1),y)
y_pred = or_lr.predict(X.reshape(-1,1))

plt.figure(figsize=(15,12))
plt.plot(X,y_pred)
plt.scatter(X,y)
print("R^2 Score is:",r2_score(y,y_pred))
plt.show()
```

```python
poly_reg = PolynomialFeatures(degree=3)
X_polly = poly_reg.fit_transform(X.reshape(-1,1))
or_lr.fit(X_polly,y)
```

```python
X_fit = np.arange(X.min(),X.max(),1).reshape(-1,1)
y_pred = or_lr.predict(poly_reg.fit_transform(X_fit))
```

```python
plt.figure(figsize=(15,12))
plt.plot(X_fit,y_pred)
plt.scatter(X,y)
print(r2_score(y,or_lr.predict(X_polly)))
```
